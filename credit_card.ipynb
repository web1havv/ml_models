```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
creditcard_dataset=pd.read_csv('/content/creditcard.csv')
```

```python
creditcard_dataset.head()
```

```python
creditcard_dataset.info()
```

```python
creditcard_dataset.tail()
```

```python
creditcard_dataset["Class"].value_counts()
```

```python
#highly unbalanced
```

```python
#Separating the data for analysis
```

```python
legit=creditcard_dataset[creditcard_dataset.Class==0]
```

```python
fraud=creditcard_dataset[creditcard_dataset.Class==1]
```

```python
fraud.shape
```

```python
legit.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
creditcard_dataset.groupby('Class').mean()
```

```python
# Now we are doing undersampling getting as number of legit transaction that of fraud
```

```python
legit_sample=legit.sample(n=492)
```

```python
# Contenating two Data frames() Used to join two data frames
```

```python
new_dataset =pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

```python

X=new_dataset.drop(columns="Class",axis=1)
```

```python
Y=new_dataset['Class']
```

```python
X.shape,Y.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.2,random_state=7400)
```

```python
model=LogisticRegression()
```

```python
X_train.shape,Y_train.shape
```

```python
model.fit(X_train,Y_train)
```

```python
train_predict=model.predict(X_train)
```

```python
accuracyscore=accuracy_score(Y_train,train_predict)
```

```python
print(accuracyscore)
```

```python
test_predict=model.predict(X_test)
```

```python
accuracyscore_train=accuracy_score(Y_test,test_predict)
```
