
Importing the Dependencies



```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection

```python
sonar_data = pd.read_csv("/content/Copy of sonar data.csv",header=None)
sonar_data
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
x=sonar_data.drop(columns=60,axis=1)
```

```python
y=sonar_data[60]
```

```python
print(x)
```

```python
print(y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,train_size=0.1,stratify=y, random_state=1)
```

```python
print(x.shape,Y_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

ISme we are comparing x train ka with original jaha pe y se ban raha hai.

```python
X_train_prediction=model.predict(X_train)
```

```python
training_data_accuracy= accuracy_score(X_train_prediction,Y_train)
```

```python
training_data_accuracy
```

```python
X_test_prediction=model.predict(X_test)
```

```python
test_data_accuracy= accuracy_score(X_test_prediction,Y_test)
```

```python
test_data_accuracy
```
